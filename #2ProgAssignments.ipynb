### Question 1:
To show that the expected value $ E[y] = \sum y \cdot P(y) = n \cdot p_i $, we first need to express $ E[y] $ in terms of the joint probability distribution $ P(y) $.

The expected value $ E[y] $ is defined as:
$ E[y] = \sum_{\text{all } y} y \cdot P(y) $

Given the joint probability distribution:
$ P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_n^{y_n} $

Now, let's express $ E[y] $ using this distribution:
$ E[y] = \sum_{\text{all } y} y_1 \cdot P(y_1, y_2, ..., y_n) $

Since $ Y_i $ equals the number of trials for which the outcome falls into cell $ i $, $ y_i $ represents the number of trials in which the outcome falls into cell $ i $. Therefore, $ y = y_1 + y_2 + ... + y_n = n $.

So, we can rewrite $ E[y] $ as:
$ E[y] = \sum_{\text{all } y} (y_1 + y_2 + ... + y_n) \cdot P(y_1, y_2, ..., y_n) $

Expanding the summation, we get:
$ E[y] = \sum_{\text{all } y} y_1 \cdot P(y_1, y_2, ..., y_n) + \sum_{\text{all } y} y_2 \cdot P(y_1, y_2, ..., y_n) + ... + \sum_{\text{all } y} y_n \cdot P(y_1, y_2, ..., y_n) $

Since $ y_1 + y_2 + ... + y_n = n $, each term in the above summation becomes:
$ E[y] = n \cdot (p_1 + p_2 + ... + p_n) = n \cdot 1 = n $

Hence, $ E[y] = n \cdot p_i $, which proves the desired result.

### Question 1:
To derive the probability distribution of the experiment described:

1. The experiment consists of $n$ identical trials.
2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$ and remains the same from trial to trial. The probability of a failure is equal to $q = (1 − p)$.
4. The trials are independent.
5. The random variable of interest is $Y$, the number of successes observed during the $n$ trials.

The probability distribution of $Y$ follows a binomial distribution, which is given by:
$ P(Y = y) = \binom{n}{y} p^y (1-p)^{n-y} $

### Question 2:
To show that the expectation of this probability distribution is $E[y] = \Sigma y \cdot P(y) = np$:

The expected value $E[Y]$ is defined as:
$ E[Y] = \sum_{\text{all } y} y \cdot P(y) $

Using the binomial distribution formula, we substitute $P(y)$:
$ E[Y] = \sum_{\text{all } y} y \cdot \binom{n}{y} p^y (1-p)^{n-y} $

Expanding and simplifying this expression yields:
$ E[Y] = np \sum_{y=0}^{n} \binom{n}{y} p^y (1-p)^{n-y} $

By the binomial theorem, this sum equals $(p + (1-p))^n$, which is equal to $1^n = 1$.

Hence, $E[Y] = np$.

Question 1: Find a general formula for the expected value of this distribution.

Question 2: A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.

$\textbf{Solution:}$
--

### Question 1:
Consider a probability distribution of $ p(y) = \frac{\lambda^y}{y!} e^{-\lambda} $.

To find a general formula for the expected value of this distribution, we need to calculate the sum of all possible outcomes weighted by their probabilities. The expected value $ E[y] $ is given by:
$ E[y] = \sum_{y=0}^{\infty} y \cdot p(y) $

Substituting the given probability distribution:
$ E[y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} $

We can simplify this expression further, recognizing that it represents the expected value of a Poisson distribution, where $ \lambda $ is the mean or expected value:
$ E[y] = \lambda $

Hence, the general formula for the expected value of this distribution is $ E[y] = \lambda $.

### Question 2:
Given that a certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard:

We can model the distribution of seedlings per square yard using a Poisson distribution with $ \lambda = 5 $. The probability that none of the ten 1-square-yard sampling regions will contain seedlings is given by:
$ P(\text{no seedlings in a region}) = p(0)^{10} $

Using the given Poisson distribution formula:
$ p(0) = \frac{5^0}{0!} e^{-5} = e^{-5} $

Substitute this value into the probability expression:
$ P(\text{no seedlings in a region}) = (e^{-5})^{10} = e^{-50} $

Hence, the probability that none of the regions will contain seedlings is $ e^{-50} $.
